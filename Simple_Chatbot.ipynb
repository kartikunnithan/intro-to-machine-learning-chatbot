```python
data={
    "hello":"Hi there!",
    "hi":"Hello!!",
    "how are you":"i'm fine thank you!",
    "name":"I am a simple chatbot",
    "bye":"Goodbye, talk soon!"
}
print("Chatbot:You can talk to me. Type 'exit' to quit\n")
while True:
  user=input("You:").lower()
  if user=="exit" or user=="bye":
    print("Chatbot:Goodbye,talk soon!")
    break
  found=False
  for key in data:
    if key== user:
      print("Chatbot:",data[key])
      found=True
      break
  if not found:
    print("Chatbot:I dont understand, am yet to be configured to respond to that")
```
